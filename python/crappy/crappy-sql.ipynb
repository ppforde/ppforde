USE crappy_gifts;

CREATE TABLE depts (
    dept_id INT NOT NULL PRIMARY KEY,
    dept_name VARCHAR(80) UNIQUE NOT NULL
);
GO

CREATE TABLE products (
    product_id INT NOT NULL PRIMARY KEY,
    product_name VARCHAR(80) UNIQUE NOT NULL,
    dept_id INT REFERENCES depts (dept_id),
    product_price FLOAT,
    product_margin FLOAT,    
    CHECK (product_price > 0.01),
    CHECK (product_margin < product_price)
);
GO

CREATE TABLE employees (
    employee_id INT IDENTITY(1,1) PRIMARY KEY,
    employee_fname VARCHAR(80),    
    employee_location CHAR(2)
);
GO

CREATE TABLE locations (
    location_id INT NOT NULL PRIMARY KEY,
    location_long VARCHAR(80),
    location_country VARCHAR(80),
    location_short VARCHAR(3),
    location_juris VARCHAR(80)
);
GO

CREATE TABLE clients (
    client_id INT NOT NULL PRIMARY KEY,
    client_sym VARCHAR(20),
    client_name VARCHAR(225),
    client_last_price FLOAT,
    client_sector VARCHAR(225)
);
GO

CREATE TABLE trans (
    src CHAR(3) CHECK (src = 'ACT' OR src = 'PLN' OR src = 'FCT'),
    employee_id INT REFERENCES employees (employee_id),
    client_id INT,
    product_id INT REFERENCES products (product_id),
    location_id INT REFERENCES locations (location_id),
    trans_qty INT,
    trans_date DATE,
    import_date DATETIME
);
GO

RETURNS INTEGER
AS
    BEGIN
        RETURN
            (SELECT CAST(ROUND((@top-@bottom) * random_number + @bottom, 0) AS INTEGER)
            FROM dbo.random_number_view)
    END

DROP VIEW IF EXISTS vtrans_details;

CREATE VIEW vtrans_details
AS
SELECT
    d.dept_name,
    p.product_name,
    p.product_price,
    p.product_margin,
    e.employee_fname,
    l.location_long,
    l.location_country,
    c.client_name,
    c.client_sector,
    t.src,
    t.trans_date,
    t.trans_qty,
    t.trans_qty * p.product_price as sales,
    t.trans_qty * (p.product_price * p.product_margin) as costs
FROM trans t
LEFT JOIN products p
ON p.product_id = t.product_id
LEFT JOIN employees e
ON e.employee_id = t.employee_id
LEFT JOIN locations l
ON l.location_id = t.location_id
LEFT JOIN clients c
ON c.client_id = t.client_id
LEFT JOIN depts d
ON p.dept_id = d.dept_id
GO

SELECT TOP 100 *
FROM vtrans_details;

CREATE VIEW vtrans_summary
AS
SELECT
    /*The day of the week as Sunday (0) to Saturday (6)*/
    src,
    DATENAME(dw, trans_date) day_name,
    DAY(trans_date) day_no,
    MONTH(trans_date) month_no,
    YEAR(trans_date) year_no,    
    product_name,
    location_long,
    location_country,
    round(sum(sales),2) total_sales,
    round(sum(costs),2) total_costs
FROM vtrans_details vt
GROUP BY
    src,
    DATENAME(dw, trans_date),
    DAY(trans_date),
    MONTH(trans_date),
    YEAR(trans_date),    
    product_name,
    location_long,
