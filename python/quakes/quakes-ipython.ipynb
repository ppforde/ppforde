```python
LINK = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
```

```python
quakes = pd.read_csv(LINK, parse_dates=['time', 'updated'])
quakes.info()
```

```python
quakes.head()
```

```python
quakes.shape
```

```python
quakes['timeframe'] = quakes['time'].apply(lambda x: str(x)[0:10])
quakes['starttime'] = quakes['time'].apply(lambda x: str(x)[0:19])
quakes['updatetime'] = quakes['updated'].apply(lambda x: str(x)[0:19])

quakes.filter(regex="time")
```

```python
quakes = quakes[quakes['mag'] > 0]
```

```python
quakes.isna().sum()
```

```python
quakes.shape
```

```python
quakes = quakes.sort_values('time')
quakes.reset_index(inplace=True, drop=True)
```

```python
from IPython.display import display
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
from datetime import datetime, timedelta
```

```python
ls = sorted(list(quakes.locationSource.unique()))
ls
```

```python
#ts = sorted(list(quakes.type.unique()))
ts = np.unique(quakes.type)
ts
```

```python
st = sorted(list(quakes.status.unique()))
st
```

```python
quakes['factors'] = pd.factorize(quakes['type'])[0]
quakes['factors'].value_counts()
```


>## Float Range Sliders

```python
@interact_manual(
    xval=widgets.FloatSlider(max=8, step=.1, readout_format='.1f'),
    yval=widgets.FloatSlider(step=.5, readout_format='.1f'),
)
def showme(xval, yval):
    
    print(f"\nResults for mag >= {xval:0.1f} and depth >= {yval:0.1f}\n")     
    
    q = quakes.query("mag >= @xval & depth >= @yval").copy()
       
    q.plot.scatter(x='mag', y='depth', c='mag', cmap='seismic', figsize=(14,4))
    
    q[['depth','mag']].plot.hist(subplots=True, layout=(1,2), figsize=(14,4))
```

>## Toggle

```python
@interact(
    zval=widgets.ToggleButtons(options=ls), 
)
def showme(zval):
    
    print(f"\nResults for {zval.upper()}\n")     
    
    q = quakes.query("locationSource == @zval").copy()   
    
    display(q)
```

>## Dropdown

```python
@interact(
    zval=widgets.Dropdown(options=ls, value='nn', description='Location'))
def showme(zval):
    
    print(f"\nResults for {zval.upper()}\n")     
    
    q = quakes.query("locationSource == @zval").copy()   
    
    display(q)
```

>## Select Multiples

```python
@interact(
    lval=widgets.SelectMultiple(options=ls, disabled=False)
)
def showme(lval):    
    
    print("For locations in", "/".join(lval))
   
    q = quakes.query("locationSource in @lval")
    
    display(q)
```

>## Radio Buttons

```python
@interact(
    sval=widgets.RadioButtons(options=st)
)
def showme(sval):   
    
    q = quakes.query("status == @sval").copy()   
    
    print(f"Results for {sval.upper() } Total= {len(q):,}")    
    
    display(q)
```

>## Date Picker

```python
q = quakes.pivot_table(
    index='timeframe',
    columns='type',
    values='mag',
    aggfunc=np.max,
    ).reset_index()
q    
```

```python
@interact(
    start=widgets.DatePicker(disabled=False, value=datetime.date(datetime.now()-timedelta(30))),
    end=widgets.DatePicker(disabled=False, value=datetime.date(datetime.now()))
)
def showme(start, end):
    
    print(f"\nResults from {start} to {end}:\n")     
    
    start_val = str(start)
    end_val = str(end)

    q = quakes.pivot_table(
        index='timeframe',
        columns='type',
        values='mag',
        aggfunc=np.max,
        ).reset_index()
        
    q[q['timeframe'].between(start_val,end_val)]\
    .plot(
        x='timeframe',
        subplots=True,
        lw=2,
        alpha=.7,
        layout=(3,3),
        rot=90,
        sharex=True,
        sharey=False,
        marker='o',
        figsize=(14,6),
    );
```

>## Text

```python
@interact(
    tval=widgets.Text(description='Place', disabled=False, value=''),    
)
def showme(tval):
    
    q = quakes[quakes['place'].str.contains(tval, case=True)].copy()
    
    print(f"\nFound {len(q):,} out of {len(quakes):,} records = {(len(q)/len(quakes))*100:.2f}%")
    
    display(q)
```

>## Int Range Sliders

```python
@interact(
    ival=widgets.IntRangeSlider(
        value=[0,len(quakes)], min=0, max=len(quakes),
        step=1,
        continuous_update=False,
        description="Index"
    ),    
)
def showme(ival):
    
    q = quakes.iloc[ival[0]:ival[-1],]
    
    #sns.histplot(data=q, x='mag');
    
    #sns.scatterplot(data=q, x='time', y='mag');    
       
    #sns.boxplot(data=q, x='locationSource', y='mag');
    
    sns.countplot(data=q, x='locationSource')
    
    #sns.lineplot(data=q, x='time', y='mag')   
    
```

>## Animation

```python
@interact(
    ival=widgets.Play(value=0, min=0, max=len(quakes), step=10, continuous_update=False),    
)
def showme(ival):
    
    print(f"\nDateframe index = {ival:}\n")         
    
    quakes.iloc[0:ival,].plot.scatter(x='mag', y='depth', c='mag', cmap='seismic', figsize=(14,4))
```
