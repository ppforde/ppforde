```python
import seaborn as sns

plt.style.use("seaborn-white")

sns.__version__
```

```python
LINK = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
```

```python
quakes = pd.read_csv(LINK, parse_dates=['time', 'updated'])
quakes.info()
```

```python
quakes.head()
```

```python
quakes.shape
```

```python
quakes['timeframe'] = quakes['time'].apply(lambda x: str(x)[0:10])
quakes['starttime'] = quakes['time'].apply(lambda x: str(x)[0:19])
quakes['updatetime'] = quakes['updated'].apply(lambda x: str(x)[0:19])

quakes.filter(regex="time")
```

```python
quakes = quakes.loc[quakes['mag']>0]
```

```python
quakes.isna().sum()
```

```python
# completed cases only
quakes = quakes.dropna(how="any")
quakes.isna().sum()
```

```python
quakes.shape
```

```python
quakes = quakes.sort_values("time").reset_index(drop=True)
quakes.nunique()
```

```python
quakes.describe()
```

```python
quakes.head()
```

```python
fig, ax = plt.subplots(figsize=(10,8), nrows=2)

sns.histplot(quakes['mag'], ax=ax[0])

sns.lineplot(data=quakes, x='timeframe', y='mag', ax=ax[1])

plt.subplots_adjust(hspace=0.2)

plt.xticks(rotation=90)

plt.show();
```

```python
f, ax = plt.subplots(1, 1, figsize=(10,4), sharex=False)

sns.countplot(data=quakes, x='locationSource', color="b", ax=ax)

ax.axhline(0, color="k")

ax.axhline(750, color="r")

sns.despine(bottom=True)

plt.tight_layout(h_pad=2)

plt.show();
```

```python
f, ax = plt.subplots(figsize=(7,6))

sns.boxplot(data=quakes, x='mag', y='type', width=0.6, color="darkblue")

sns.stripplot(data=quakes, x='mag', y='type', size=4, color="lightblue", alpha=0.2, linewidth=0)

ax.xaxis.grid(True)

ax.set(ylabel="")

sns.despine(trim=True, left=True)

plt.show();
```

```python
g = sns.JointGrid(data=quakes, x='gap', y='depth', space=0.5)
g.plot_joint(sns.kdeplot, fill=True, thresh=0, levels=100, cmap="coolwarm")
g.plot_marginals(sns.histplot, color="blue", alpha=1, bins=25)

plt.show();
```

```python
g = sns.jointplot(data=quakes, kind="kde", x='gap', y='depth')
g.plot_joint(plt.scatter, c="w", s=30, linewidth=.5, marker="+")
g.ax_joint.collections[0].set_alpha(0)
g.set_axis_labels("$Gap$", "$Depth$")

plt.show();
```

```python
q = quakes.groupby('magSource', as_index=False).mean()
q
```

```python
q.columns[:8]
```

```python
color_groups = ['Blues','Greens','Reds','Purples','Oranges']

for i, Y in enumerate(q.columns[3:8]):
    f, ax = plt.subplots(figsize=(10,3))
    ax = sns.barplot(data=q, x='magSource', y=q[Y], capsize=None, palette=color_groups[i])
```

```python
fig, ax = plt.subplots(figsize=(10,4))

sns.set_color_codes("dark")
sns.barplot(data=q, x='magError', y='magSource', label="Mag Error", color="b")

sns.set_color_codes("muted")
sns.barplot(data=q, x='mag', y='magSource', label="Mag", color="b", alpha=0.2)

ax.legend(ncol=2, loc="lower right", frameon=True)
ax.set(xlim=(0, q['mag'].max() * 1.5), ylabel="", xlabel="Mag vs Mag Error")

sns.despine(left=True, bottom=True)

plt.show();
```

```python
f, ax = plt.subplots(figsize=(6,6))
cmap = sns.cubehelix_palette(as_cmap=True, dark=0, light=1, reverse=True)
sns.kdeplot(data=quakes, x='mag', y='gap', cmap=cmap, n_levels=60, shade=True)

plt.show();
```

```python
f, ax = plt.subplots(figsize=(8,8))

ax = sns.kdeplot(data=quakes.query("magSource=='pr'"), x='depth', y='gap', cmap="Reds", shade=True, thresh=0.05)
ax = sns.kdeplot(data=quakes.query("magSource=='mb'"), x='depth', y='gap', cmap="Blues", shade=True, thresh=0.05)
ax = sns.kdeplot(data=quakes.query("magSource=='uu'"), x='depth', y='gap', cmap="Greens", shade=True, thresh=0.05)

red = sns.color_palette("Reds")[-2]
blue = sns.color_palette("Blues")[-2]
green = sns.color_palette("Greens")[-2]

ax.text(25,50, "mb", size=16, color=blue)
ax.text(75,300, "pr", size=16, color=red)
ax.text(25,25, "uu", size=16, color=green)

plt.show();
```
