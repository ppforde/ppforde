```python
LINK = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
```

```python
quakes = pd.read_csv(LINK, parse_dates=['time', 'updated'])
quakes.info()
```

```python
quakes.head()
```

```python
quakes.shape
```

```python
quakes['timeframe'] = quakes['time'].apply(lambda x: str(x)[0:10])
quakes['starttime'] = quakes['time'].apply(lambda x: str(x)[0:19])
quakes['updatetime'] = quakes['updated'].apply(lambda x: str(x)[0:19])

quakes.filter(regex="time")
```

```python
quakes.isna().sum()
```

```python
#completed cases only
quakes = quakes.dropna(how='any')
quakes.isna().sum()
```

```python
quakes.shape
```

```python
quakes = quakes.sort_values('time')
quakes.reset_index(inplace=True, drop=True)
```

```python
quakes.nunique()
```

```python
quakes.describe()
```

## Bokeh Plots

```python
q = quakes.groupby('magType', as_index=False)['mag'].mean()
q
```

```python
q.plot_bokeh(
    kind='pie',
    x='magType',
    y='mag',
    line_color='white',
    line_width=5,
);
```

```python
quakes.plot_bokeh(
    kind='scatter',
    x='mag',
    y='depth',
    category='type',
    legend="top_left",
    line_width=0,
    alpha=.4,
    disable_scientific_axes='xy',
);
```

```python
t = set(quakes.type)
t
```

```python
plots = []

colors = ['red', 'orange', 'yellow', 'green', 'blue', 'violet', 'magenta']

colors = ['#882426', '#de7a22', '#c29545', '#a2c523', '#004445', '#5d535e', '#2d4262']

for i, x in enumerate(sorted(t)):    
    p = quakes.query("type == @x")\
    .plot_bokeh(
        kind='scatter',
        x='mag',
        y='depth',
        category='magType',
        alpha=.5,
        color=colors[i],
        line_width=0,
        title=x.capitalize(),
        disable_scientific_axes="xy",
        hovertool=False,
        legend=False,
        show_figure=False
    );
    plots.append(p)
    
pandas_bokeh.plot_grid([plots[:3], plots[3:]], plot_width=300, plot_height=300);
```

```python
tile_providers = [
    'CARTODBPOSITRON_RETINA',
    'CARTODBPOSITRON',
    'STAMEN_TERRAIN',
    'STAMEN_TERRAIN_RETINA',
    'STAMEN_TONER',
    'STAMEN_TONER_BACKGROUND',
    'STAMEN_TONER_LABELS'
]
```

```python
quakes.query("mag > 0").plot_bokeh.map(
    x="longitude",
    y="latitude",
    hovertool_string="""Place: @{place},<br> Magnitude: @{mag}""",
    size="mag",
    colormap='Inferno',
    figsize=(800, 600),
    legend=None,
    title="Earthquakes"
);
```

```python
quakes.query("mag > 0").plot_bokeh.map(
    x="longitude",
    y="latitude",
    hovertool_string="""Place: @{place},<br> Magnitude: @{mag}""",
    tile_provider="STAMEN_TERRAIN_RETINA",
    size="mag",
    colormap='Inferno',
    number_format='0.00',
    legend=None,
    figsize=(800, 600),
    title="Earthquakes"
);
```

```python
q = quakes.pivot_table(
    index='timeframe',
    columns='type',
    values='mag',
    fill_value=0,
    aggfunc=np.max,
)

q
```

```python
q.plot_bokeh(
    kind='line',
    sizing_mode='scale_width',
    plot_data_points=True,
    plot_data_points_size=7,
    rangetool=True,
    legend='bottom_right',
    marker='square',
    figsize=(900,400),
);
```

```python
q.plot_bokeh(
    kind='line',
    y=['earthquake','explosion'],
    sizing_mode='scale_width',
    plot_data_points=True,
    plot_data_points_size=7,
    rangetool=True,
    marker='square',
    legend='bottom_right',
    figsize=(900,400),
);
```
