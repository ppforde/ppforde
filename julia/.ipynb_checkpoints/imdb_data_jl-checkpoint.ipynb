```julia
using CSV, DataFrames, GZip, HTTP
```

```julia
URL = "https://datasets.imdbws.com/title.basics.tsv.gz"
```

```julia
;wget https://datasets.imdbws.com/title.basics.tsv.gz
```

```julia
;gzip -df title.basics.tsv.gz
```

```julia
file = download(URL)
```

```julia
ENV["LINES"] = 25
```

```julia
ENV["COLUMNS"] = 200
```

```julia
df = DataFrame(CSV.File(GZip.open(file), delim="\t", header=true, missingstring="\\N", quotechar=',', closequotechar='"'))
```

```julia
typeof(df)
```

```julia
size(df)
```

```julia
names(df)
```

```julia
unique(df.isAdult)
```

```julia
df = filter(x -> x.isAdult==0, df)
```

```julia
df = filter(x -> !ismissing(x.startYear), df)
```

```julia
df = filter(x -> !ismissing(x.runtimeMinutes), df)
```

```julia
df = filter(x -> !ismissing(x.genres), df)
```

```julia
first(df, 5)
```

```julia
last(df, 5)
```

```julia
years = 1995:2000

filter(x -> in(x.startYear, years), df)
```

```julia
filter(x -> startswith(x.primaryTitle, "Star"), df)
```

```julia
filter(x -> endswith(x.primaryTitle, "Star"), df)
```

```julia
filter(x -> contains(x.primaryTitle, "Star Trek"), df)
```

```julia
filter(x -> x.runtimeMinutes > 300, df)
```

```julia
site = "https://www.imdb.com/chart/top"
```

```julia
r = HTTP.request("GET", site)
data = String(r.body);
```

```julia
occursin("Shawshank", data)
```

```julia
length(data)
```

```julia
mregex = r"(tt\d+)"
```

```julia
match(mregex, data)
```

```julia
findall(mregex, data)
```

```julia
tconst = [data[d] for d in findall(mregex, data)]
```

```julia
tconst = unique(tconst)
```

```julia
filter(x -> in(x.tconst, tconst), df)
```
